```
from keras.models import Sequential
from keras.layers import*
```

```
#NN with Single Neuron
model=Sequential()
model.add(Dense(6,input_shape=(2,),activation='tanh'))#Output Neuron we can use softmax 
model.add(Dense(4,activation='tanh'))#Output Neuron we can use softmax 

model.add(Dense(2,activation='softmax'))#Output Neuron we can use softmax 
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=["accuracy"])
model.summary()
```

```
#Train
from keras.utils import to_categorical 
Y_oht= to_categorical(Y)
print(Y_oht.shape)

model.fit(X,Y_oht,epochs=500)
```

```
from keras.preprocessing import image
```

```
img=image.load_img("Train/Balbasaur/425.jpg",target_size=(100,100))#remove target size for better quality
img=image.img_to_array(img)/255.0
#import matplotlib.pyplot as plt
plt.imshow(img)
```

```
print(img.shape)
```

```
img[:,:,1:]=0
plt.imshow(img)
```

```
# CNN Model
from keras.models import Sequential
from keras.layers import *
model = Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(100,100,3)))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
          
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
          
model.add(GlobalAveragePooling2D())
#Classifier 
model.add(Dense(64,activation='relu'))
model.add(Dense(3,activation='softmax'))

model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
model.summary()
```

```
#Data Loader with Augumentation
train_datagen = image.ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,rescale=1.0/255,
    horizontal_flip=True)
test_datagen= image.ImageDataGenerator(rescale=1.0/255)
```

```
train_loader=train_datagen.flow_from_directory('Train',batch_size=32,class_mode='categorical',shuffle=True)
val_loader=test_datagen.flow_from_directory('Test',batch_size=32,class_mode='categorical',shuffle=True)
```

```
for x,y in train_loader:
  print(x.shape,y.shape)
  for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x[i])
    plt.axis("off")
  plt.show()
  break
plt.savefig("Augumentation and Generator")
```

```
for x,y in val_loader:
  print(x.shape,y.shape)
  for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x[i])
    plt.axis("off")
  plt.show()
  break
plt.savefig("Augumentation and Generator_Test Data")
```

```
hist=model.fit_generator(train_loader,epochs=50,steps_per_epoch=10,shuffle=True,validation_data=val_loader,validation_steps=2)
```

```
img=img.reshape(1,100,100,3)

pred=np.argmax(model.predict(img),axis=1)[0]
name2idx=train_loader.class_indices
idx2name=dict(k,name2idx[k],k) for k in name2idx.keys())
print(idx2name)
print(name2idx)
output=is=dx2name[pred]
print(output)
```
